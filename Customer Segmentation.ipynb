```python
#clustering the data using k-means /elbow technique
X1=data.loc[:, ["Age","Spending Score(1-100)"]].values
from sklearn.cluster import KMeans
wcss=[]
for k in range(1,11):
    kmeans=KMeans(n_clusters=k, init="k-means++")
    kmeans.fit(X1)
    wcss.append(kmeans.inertia_)
plt.figure(figsize = (12,6))
plt.grid()
plt.plot(range(1,11),wcss, linewidth=2, color="red", marker="8")
plt.xlabel("K Value")
plt.ylabel("WCSS")
plt.show()
```

```python
kmeans= KMeans(n_clusters=4)

label= kmeans.fit_predict(X1)

print(label)
```

```python
print(kmeans.cluster_centers_)
```

```python
#plotting centroids of the clusters between age and spending score
plt.scatter(X1[:,0], X1[:,1], c=kmeans.labels_, cmap='rainbow')
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1],color='black')
plt.title('Clusters of customer')
plt.xlabel('Age')
plt.ylabel('Spending Score(1-100)')
plt.show()
```

```python
#clustering data btw annual income and spending score
X2=data.loc[:, ["Annual Income(k$)","Spending Score(1-100)"]].values
from sklearn.cluster import KMeans
wcss=[]
for k in range(1,11):
    kmeans=KMeans(n_clusters=k, init="k-means++")
    kmeans.fit(X2)
    wcss.append(kmeans.inertia_)
plt.figure(figsize = (12,6))
plt.grid()
plt.plot(range(1,11),wcss, linewidth=2, color="red", marker="8")
plt.xlabel("K Value")
plt.ylabel("WCSS")
plt.show()
```

```python
kmeans= KMeans(n_clusters=5)

label= kmeans.fit_predict(X2)

print(label)
```

```python
# printing their centroid co ods
print(kmeans.cluster_centers_)
```

```python
#plotting centroids of the clusters between annual income and spending score
plt.scatter(X2[:,0], X2[:,1], c=kmeans.labels_, cmap='rainbow')
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1],color='black')
plt.title('Clusters of customer')
plt.xlabel('Annual Income(k$)')
plt.ylabel('Spending Score(1-100)')
plt.show()
```
